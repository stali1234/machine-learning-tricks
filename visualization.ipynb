```python
import io
import pandas as pd
train_df = pd.read_csv(io.BytesIO(uploaded['movie_metadata.csv']))
```

```python
import numpy as np
import matplotlib.pyplot as plt
train_df.head()
```

```python
train_df.columns
```

```python
titles = list(train_df.get('movie_title'))
print(type(titles))
print(titles[0])
```

```python
freq_titles = {}
for title in titles:
  length = len(title)

  if freq_titles.get(length) is None:
    freq_titles[length] = 1
  else:
    freq_titles[length] +=1  
```

```python
x = np.array(list(freq_titles.keys()))
y = np.array(list(freq_titles.values()))
plt.scatter(x,y)
plt.xlabel("length of movie title")
plt.ylabel("freq count")

plt.show()
```
