```python
titles = list(train_df.get('movie_title'))
titles
```

```python
freq_title = {}
for t in titles:
  length = len(t)
  try:
    freq_title[length] += 1
  except:
    freq_title[length] = 1 
  print(freq_title)
```

```python
x  = np.array(list(freq_title.keys()))
y = np.array(list(freq_title.values()))
plt.scatter(x,y)
plt.xlabel("movie title len")
plt.ylabel("no of movies")
plt.show()

```
