```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line plot
x = np.arange(10)
y = x**2
y2 = x**3
print(x)
print(y)
print()
```

```python
plt.style.use('seaborn')
plt.plot(x,y2 ,color='green',label="apple")
plt.plot(x,y,color='red',label='kiwi')
plt.xlabel("time")
plt.ylabel("price")
plt.legend()

plt.show()
```

```python
prices = np.array([1,2,3,3,4])**2
print(prices)
plt.plot(prices)
plt.show()
```

```python
#saccter plot
plt.style.use('seaborn')
plt.scatter(x,y2 ,color='green',label="apple")
plt.scatter(x,y,color='red',label='kiwi')
plt.xlabel("time")
plt.ylabel("price  of fruits over ")
plt.legend()

plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])
plt.bar([0,1,2],[10,20,30],width=0.5,label="current year",tick_label=["gold","platinum","silver"])
plt.title("metal price comparison")
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5 , label="next year")
plt.xlabel("metal")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
#pie chart
subjects = "maths","chem","pyhics"
weightage = [20,40,56]
plt.pie(weightage ,labels=subjects , explode = (0,0,1)) #highlights
plt.show()
```

```python
#histogram

xsn = np.random.randn(100)
sigma = 5
u = 60
X = np.round(xsn*sigma + u)
print(X)
```

```python
plt.style.use('seaborn')
plt.hist(X)
plt.show()
```
